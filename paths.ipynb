```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('paths.txt', header=None).T
df['mean'] = df.mean(axis=1)
```

```python
for x in df.columns:
    plt.plot(df.index, df[x])
```

```python
df['mean_long'] = df['mean'].ewm(alpha=2/(500+1)).mean()
df['mean_short'] = df['mean'].ewm(alpha=2/(50+1)).mean()
# plt.plot(df.index, df['mean'])
plt.plot(df.index, df['mean_long'])
plt.plot(df.index, df['mean_short'])
plt.legend()
```
